### Import dependencies

We will import `tensorflow` and `numpy`.

```python
import tensorflow as tf
import numpy as np
```

```python
tf.__version__
```

### Create a NumPy array

A sequence of Square Numbers; 1, 4, 9, 16, 25, 36.

```python
numpy_array = np.array([1, 4, 9, 16, 25, 36])
```

Print out the `numpy_array`.

```python
print(numpy_array)
```

Observe the `type` and the data type object (`dtype`) of the `numpy_array`.

```python
type(numpy_array)
```

```python
numpy_array.dtype
```

The `type` of the `numpy_array` is `numpy.ndarray` and the `dtype` is `int64`.

### Create a TensorFlow tensor from this `numpy_array`

```python
tensorflow_tensor = tf.convert_to_tensor(numpy_array)
```

Print out the `tensorflow_tensor`.

```python
print(tensorflow_tensor)
```

Observe the `type` and the data type object (`dtype`) of the `tensorflow_tensor`.

```python
type(tensorflow_tensor)
```

```python
tensorflow_tensor.dtype
```

The `type` of the `tensorflow_tensor` is `tensorflow.python.framework.ops.EagerTensor` and the `dtype` is `tf.int64`.
