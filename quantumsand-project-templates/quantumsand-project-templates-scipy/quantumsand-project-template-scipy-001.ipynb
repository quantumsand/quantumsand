### Import dependencies

We will import `scipy` and `numpy`.

```python
from scipy.integrate import quad
import numpy as np
```

### Use a lambda function to integrate The Gaussian Integral

$$ \int_{-\infty}^\infty e^{-x^2}\,dx = \sqrt{\pi} $$

We can express this integral as a Python lambda function using the following syntax: 

`lambda x: np.exp(-x ** 2)`.

* `np.exp()` calculates the exponential of all elements in the input array.
* `-x ** 2` is negative x to the second power.

Using `quad()` within `scipy.integrate` allows us to integrate a function of one variable between two points. We can use the following arguments;

* `func` is a Python function or method to integrate, `lambda x: np.exp(-x ** 2)`.
* `a` is the lower limit of integration, negative infinity, `-np.inf`.
* `b` is the upper limit of integration, positive infinity, `np.inf`.

We can save the result of `quad()` to `the_gaussian_integral`.

We can also calculate $ \sqrt{\pi} $ using `np.sqrt()` and `np.pi` into `the_square_root_of_pi`.

```python
the_gaussian_integral = quad(lambda x: np.exp(-x ** 2), -np.inf, np.inf)
the_square_root_of_pi = np.sqrt(np.pi)

print(f"The Gaussian Integral is: {the_gaussian_integral}")
print(f"The square root of pi is: {the_square_root_of_pi}")
```

You can see that the integration of The Gaussian Integral provides the correct result.
