### Import dependencies

We will import `tqdm` and `time`.

```python
from tqdm import tqdm
import time
```

### Wrap a for loop iterable with `tqdm()`

Using Python `range()` we can also specify the argument;

* `end` is 50.

The output of `range(50)` iterates over the numbers from 0 to 49.

Wrapping the `range(50)` inside `tqdm()` is essentially `tqdm(iterable)` and is sufficient to show a progress meter.

To deliberately slow down this for loop, we can use `sleep()` from Python time and set the argument;

* `secs` to 0.1.

We will deliberately pause; suspend execution of the calling thread; for 0.1 seconds within each of the 50 iterations through the for loop.

```python
for i in tqdm(range(50)):
    time.sleep(0.1)
```

If you are looking at this notebook statically inside a browser or VS Code you will just see the 100% completed progress bar.

If you run this block of Python code interactively, you can see the animated progress bar whilst the for loop iterates.

If you modify the numbers in this example;

* `end` within `range()`, the number of iterations in the for loop.
* `secs` within `sleep()`, the number of seconds to suspend the calling thread.

You can see that the progress bar will take a different amount of time to complete.

An example scenario for using a tqdm progress bar is during the training process of an AI model.
