### Import dependencies

We will import `nibabel` and `matplotlib`.

```python
import nibabel as nib
```

```python
import matplotlib.pyplot as plt
```

### Read the MRI scans NIfTI file 

We will use `load()` within NiBabel to load the MRI scans NIfTI file.

```python
mri_scans_nifti = nib.load("../quantumsand-project-templates-resources/nifti_mri_scans.nii.gz")
```

Print out the `mri_scans_nifti` file.

```python
print(mri_scans_nifti)
```

### Extract and render the MRI NIfTI scan images

We will use `get_fdata()` within NiBabel to obtain the image data.

```python
mri_scans_nifti_image_data = mri_scans_nifti.get_fdata()
```

We are going to display 25 of the MRI scans within the same plot as subplots. Within Matplotlib we can use `subplots()` to specify what we want. We will call `subplots()` with the following arguments;

* `nrows` is 5.
* `ncols` is 5.
* `figsize` is a keyword argument, we set to `=(15, 15)`. The (width, height) in inches.

`subplots()` returns both `fig` and `ax`, the `Figure` and the `Axes`.

* `Figure` is the top level container for all the plot elements.
* `Axes` is an object that encapsulates all of the elements of an individual subplot in a figure.

Using Python nested for loops, we can “loop inside a loop”. Also, using Python `range()` we can also specify the argument;

* `end` is 5.

The output of `range(5)` iterates over the numbers from 0 to 4.

We can also count the iterations through the nested for loops which we can call `slice_index` and then use this to obtain a 2d slice from within the NIfTI file.

It is very important to recognise that the viewing plane of the NIfTI file determines how to correctly obtain the slices. In this example scenario we will slice the data using the following syntax: `mri_scans_nifti_image_data[:,:,slice_index]`.

To populate the correct subplot we can use `imshow()` within Matplotlib to display the data as an image.

If you wish to, you can switch the `axis()` to `off` using `ax[i][j].axis('off')`.

```python
fig, ax = plt.subplots(5, 5, figsize=(15, 15))

slice_index = 0
for i in range(5):
    for j in range(5):
        ax[i][j].imshow(mri_scans_nifti_image_data[:,:,slice_index], cmap="gray")
        slice_index+=1
```

You can see the MRI NIfTI scan images.
