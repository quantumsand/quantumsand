### Converting a directory of DICOM MRI scans to NIfTI format

The NIfTI (Neuroimaging Informatics Technology Initiative) format can be used for multi-dimensional data from MRI, fMRI and CT scans. 

We will use `dicom_series_to_nifti()` within dicom2nifti to generate `nifti_mri_scans.nii.gz`

```python
path_to_dicom_mri_scans = "../quantumsand-project-templates-resources/SE000001"
generated_nifti_mri_scans_file = "../quantumsand-project-templates-resources/nifti_mri_scans.nii.gz"
```

```python
dicom2nifti.dicom_series_to_nifti(path_to_dicom_mri_scans, generated_nifti_mri_scans_file)
```

This should generate the NIfTI format file, which is already available within the `quantumsand-project-templates-resources` directory.
