### Import dependencies

We will import `seaborn`.

```python
import seaborn as sns
```

### Load the Palmer Penguins dataset into a pandas `DataFrame`

Load the `pandas_dataframe_penguins_dataset` using `load_dataset()` from Seaborn with the arguments;

* `name` is `"penguins"`.
* `data_home` is the directory in which to cache the data, `"../quantumsand-project-templates-resources"`.

The result of `load_dataset()` in this scenario is the already downloaded `"penguins.csv"` within the `quantumsand-project-templates-resources` directory.

By specifying this existing `data_home` path with the precached `"penguins.csv"` dataset, it should work fine without requiring an internet connection.

Using Python `print()` we can observe the `pandas_dataframe_penguins_dataset`.

```python
pandas_dataframe_penguins_dataset = sns.load_dataset("penguins", data_home="../quantumsand-project-templates-resources")
```

```python
print(pandas_dataframe_penguins_dataset)
```

### Render the Palmer Penguins dataset using Seaborn

Seaborn builds on top of Matplotlib to produce informative plots.

The plotting functions operate on a pandas `DataFrame` or array containing whole datasets.

Note that we do not need to explictly `import matplotlib` or `import pandas`.

However, Seaborn is using both Matplotlib and pandas as dependencies.

Apply the default theme using `set_theme()` from Seaborn.

```python
sns.set_theme()
```

Using `jointplot()` from Seaborn, we can draw a plot of two variables with bivariate and univariate graphs.

With these arguments;

* `data` is `pandas_dataframe_penguins_dataset`.
* `x` is `"flipper_length_mm"`.
* `y` is `"bill_length_mm"`.
* `hue` is `"species"`.

```python
sns.jointplot(data=pandas_dataframe_penguins_dataset, x="flipper_length_mm", y="bill_length_mm", hue="species")
```

Using `pairplot()` from Seaborn, we can plot pairwise relationships in a dataset.

With these arguments;

* `data` is `pandas_dataframe_penguins_dataset`.
* `hue` is `"species"`.

```python
sns.pairplot(data=pandas_dataframe_penguins_dataset, hue="species")
```

You can see how Seaborn plots provide useful and informative summaries of the Palmer Penguins dataset.
