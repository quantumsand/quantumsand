## Quantum Sand Project Template GeoPandas 001

### Import dependencies

We will import `geopandas`.

```python
import geopandas as gpd
```

### Import Natural Earth vector dataset into a GeoPandas `GeoDataFrame`

We will use GeoPandas to import the Natural Earth vector data.

We will use `read_file()` within GeoPandas to read a zip file into a GeoPandas `GeoDataFrame`. We will call `read_file()` with the following argument;

* `filename` in this scenario, is the filename of the zip file.

A GeoPandas `GeoDataFrame` can be used for working with geospatial data. It is an extension of standard pandas `DataFrames`. It allows us to handle geospatial geometry like points, lines, and polygons.

```python
natural_earth_vector_countries_zip = "../quantumsand-project-templates-resources/ne_110m_admin_0_countries.zip"
```

```python
geopandas_geodataframe_countries = gpd.read_file(natural_earth_vector_countries_zip)
```

### Print and render the contents of the GeoPandas `GeoDataFrame`

We can print out the `geopandas_geodataframe_countries`.

```python
print(geopandas_geodataframe_countries)
```

You can see the contents of `geopandas_geodataframe_countries`.

We can also render out the `geopandas_geodataframe_countries` using `plot()`. This will generate a plot of a `GeoDataFrame` using Matplotlib. In this example scenario we do not need to explicity import Matplotlib. We will provide `plot()` with the following arguments;

* `column` is "SOV_A3", which is the 3-letter country code.
* `edgecolor` is "black" and is a style option to be passed on to the actual plot function.

```python
geopandas_geodataframe_countries.plot(column="SOV_A3", edgecolor="black")
```

You can see the `plot()` of `geopandas_geodataframe_countries`.
