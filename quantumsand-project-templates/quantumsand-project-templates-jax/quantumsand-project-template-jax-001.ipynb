### Import dependencies

We will import `jax` and `numpy`.

```python
import jax
import jax.numpy as jnp
import numpy as np
```

```python
jax.__version__
```

### Create a NumPy array

A sequence of Square Numbers; 1, 4, 9, 16, 25, 36.

```python
numpy_array = np.array([1, 4, 9, 16, 25, 36])
```

Print out the `numpy_array`.

```python
print(numpy_array)
```

Observe the `type` and the data type object (`dtype`) of the `numpy_array`.

```python
type(numpy_array)
```

```python
numpy_array.dtype
```

The `type` of the `numpy_array` is `numpy.ndarray` and the `dtype` is `int64`.

### Create a JAX array from this `numpy_array`

```python
jax_array = jnp.array(numpy_array)
```

Print out the `jax_array`.

```python
print(jax_array)
```

Observe the `type` and the data type object (`dtype`) of the `jax_array`.

```python
type(jax_array)
```

```python
jax_array.dtype
```

The `type` of the `jax_array` is `jaxlib._jax.ArrayImpl` and the `dtype` is `int32`.
