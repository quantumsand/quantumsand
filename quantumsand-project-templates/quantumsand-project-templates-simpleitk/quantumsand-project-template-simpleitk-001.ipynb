### Import dependencies

We will import `SimpleITK` and `matplotlib`.

```python
import SimpleITK as sitk
```

```python
import matplotlib.pyplot as plt
```

### Obtain the filenames of a directory of DICOM MRI scans

We will use `ImageSeriesReader.GetGDCMSeriesFileNames()` within SimpleITK to automatically obtain all of the filenames for a particular directory path.

```python
path_to_dicom_mri_scans = "../quantumsand-project-templates-resources/SE000001"
dicom_mri_scan_filenames = sitk.ImageSeriesReader.GetGDCMSeriesFileNames(path_to_dicom_mri_scans)
```

Print out the `dicom_mri_scan_filenames`.

```python
print(dicom_mri_scan_filenames)
```

### Extract and render the DICOM MRI scan images

We will use `ImageSeriesReader()` within SimpleITK to obtain the image data. We can pass the `dicom_mri_scan_filenames` to `SetFileNames()`. Finally, calling `Execute()` will invoke the actual image reading.

```python
simple_itk_image_series_reader = sitk.ImageSeriesReader()
simple_itk_image_series_reader.SetFileNames(dicom_mri_scan_filenames)

dicom_mri_scan_image_data = simple_itk_image_series_reader.Execute()
```

We are going to convert the `dicom_mri_scan_image_data` into a NumPy array using SimpleITK `GetArrayFromImage()`.

```python
dicom_mri_scan_numpy_array = sitk.GetArrayFromImage(dicom_mri_scan_image_data)
```

Observe the `type` of the `dicom_mri_scan_numpy_array`.

```python
type(dicom_mri_scan_numpy_array)
```

You can see that the `type` of `dicom_mri_scan_numpy_array` is `numpy.ndarray`.

We are going to display 25 of the MRI scans within the same plot as subplots. Within Matplotlib we can use `subplots()` to specify what we want. We will call `subplots()` with the following arguments;

* `nrows` is 5.
* `ncols` is 5.
* `figsize` is a keyword argument, we set to `=(15, 15)`. The (width, height) in inches.

`subplots()` returns both `fig` and `ax`, the `Figure` and the `Axes`.

* `Figure` is the top level container for all the plot elements.
* `Axes` is an object that encapsulates all of the elements of an individual subplot in a figure.

Using Python nested for loops, we can “loop inside a loop”. Also, using Python `range()` we can also specify the argument;

* `end` is 5.

The output of `range(5)` iterates over the numbers from 0 to 4.

We can also count the iterations through the nested for loops which we can call `slice_index` and then use this to obtain a 2d DICOM MRI slice from the initial directory of scan images.

To populate the correct subplot we can use `imshow()` within Matplotlib to display the data as an image.

If you wish to, you can switch the `axis()` to `off` using `ax[i][j].axis('off')`.

```python
fig, ax = plt.subplots(5, 5, figsize=(15, 15))

slice_index = 0
for i in range(5):
    for j in range(5):
        ax[i][j].imshow(dicom_mri_scan_numpy_array[slice_index], cmap="gray")
        slice_index+=1
```

You can see the DICOM MRI scan images.
