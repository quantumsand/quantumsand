### Import dependencies

We will import `numpy` and `time`.

```python
import numpy as np
```

```python
import time
```

### Create an `ndarray` of 1 million random values with NumPy

An `ndarray` is a fast, memory-efficient, homogeneous container within NumPy. A homogeneous container contains items or elements of the exact same type.

```python
random_values_ndarray_size = 1000000
```

```python
numpy_random_float_values = np.random.random(random_values_ndarray_size)
```

Observe the `type` of the `numpy_random_float_values`.

```python
type(numpy_random_float_values)
```

The `type` of the `numpy_random_float_values` is `numpy.ndarray`.

Print out the `numpy_random_float_values`.

```python
print(numpy_random_float_values)
```

### Delta time $\Delta t$ for Python `sum()`

Measure the time taken to run Python `sum()` over all items within `numpy_random_float_values`.

```python
python_sum_start_time = time.time()
python_sum = sum(numpy_random_float_values)
python_sum_end_time = time.time()

python_sum_delta_time = python_sum_end_time - python_sum_start_time

print(f"Python sum() result: {python_sum}")
print(f"Python sum() delta time: {python_sum_delta_time:.3f}s")
```

### Delta time $\Delta t$ for NumPy `sum()`

Measure the time taken to run NumPy `sum()` over all items within `numpy_random_float_values`.

```python
numpy_sum_start_time = time.time()
numpy_sum = np.sum(numpy_random_float_values)
numpy_sum_end_time = time.time()

numpy_sum_delta_time = numpy_sum_end_time - numpy_sum_start_time

print(f"Numpy sum() result: {numpy_sum}")
print(f"NumPy sum() delta time: {numpy_sum_delta_time:.3f}s")
```

### Delta time  $\Delta t$ ratio between Python `sum()` and NumPy `sum()`

Calculate the ratio between the time taken to run Python `sum()` and the time taken to run NumPy `sum()`.

```python
delta_time_ratio = python_sum_delta_time / numpy_sum_delta_time
print(delta_time_ratio)
```

You can see in this example scenario, it is much faster to use NumPy `sum()` than Python `sum()`.
