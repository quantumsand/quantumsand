### Download the FashionMNIST dataset

If you are just looking at this notebook statically in a browser or VS Code you can just read this notebook without running the Python code.

You need to be able to locally download the FashionMNIST dataset to the local filesystem to run the Python code later in this notebook.

If you have not already downloaded the FashionMNIST dataset into `~/Datasets/FashionMNIST`, we can use `datasets.FashionMNIST()` from `torchvision` to obtain this data with the following arguments;

* `root`, the root directory of the dataset, is `~/Datasets/FashionMNIST`.
* `train`, a boolean, true or false, *is this training data?*; `True` or *is this test data?*; `False`.
* `download`, a boolean, true or false, if `True`, we download the dataset from the internet.
* `transform`, a function/transform that takes in a PIL image and returns a transformed version, is `v2.Compose()`.

For `download` set to `True` it is worth noting that if the dataset is already downloaded, it is not downloaded again.

`v2.Compose()` composes several transforms together. Let us examine this line;
```python
transform=v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)])
```

* `ToImage()` will convert a tensor, ndarray, or PIL Image to `Image`; this does not scale values.
* `ToDtype(torch.float32, scale=True)` converts the input to a specific dtype (data type object), `torch.float32`.

Note that PyTorch `Image` is a `torch.Tensor` subclass for images with shape.

Also note that within `ToDtype()` setting `scale` to `True` will scale the values for images or videos.

The outcome will be the following data;

* `fashion_mnist_training_data`, where `FashionMNIST()` `train` is `True`.
* `fashion_mnist_test_data`, where `FashionMNIST()` `train` is `False.`

````python
fashion_mnist_training_data = datasets.FashionMNIST(
    root="~/Datasets/FashionMNIST",
    train=True,
    download=True,
    transform=v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)]),
)

fashion_mnist_test_data = datasets.FashionMNIST(
    root="~/Datasets/FashionMNIST",
    train=False,
    download=True,
    transform=v2.Compose([v2.ToImage(), v2.ToDtype(torch.float32, scale=True)]),
)
````

Print out the `fashion_mnist_training_data` and the `fashion_mnist_test_data`.

````python
print("fashion_mnist_training_data")
print(fashion_mnist_training_data)

print()
print("fashion_mnist_test_data")
print(fashion_mnist_test_data)
````

### AI acceleration via silicon chips designed for AI workloads

Depending on your device you may or may not have AI acceleration via silicon chips designed for AI workloads to speed up PyTorch.

You may have one of the following:

* macOS uses `"mps"`.
* Linux or Windows Subsystem for Linux (WSL) may use `"cuda"`.

If no silicon chips for AI workloads are found, PyTorch can use the `"cpu"` instead.

To obtain the correct accelerator or fallback to the cpu we can use the following line;

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
```

The Python `if` statement runs `accelerator.is_available()` from PyTorch which returns `True` or `False`;

* `accelerator.is_available()` returns `True` if all of the required drivers are available and at least one device is visible.
* `accelerator.is_available()` returns `False` if no compatible accelerators are visible to PyTorch.

If `True` is returned then `device` will be the return value of `torch.accelerator.current_accelerator().type`.

Something like `"mps"` or `"cuda"` will be returned from `torch.accelerator.current_accelerator().type`.

If `False` is returned then`device` will be `"cpu"`.

In this example we either set `device` to `torch.accelerator.current_accelerator().type` or `"cpu"`.

We can `print()` the `device` we are using like so;

```python
print(f"Using {device} device")
```

````python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
print(f"Using {device} device")
print(F"torch.accelerator.current_accelerator().type {torch.accelerator.current_accelerator().type}")
````

### Define a neural network in PyTorch

We can use a Python `class` for our `NeuralNetwork`. A Python `class` provides a means of bundling data and functionality together.

PyTorch `nn.Module` is the base class for all neural network modules.

Our AI models should also subclass this class.

We initialise the neural network layers in `__init__`.

Also, our `nn.Module` subclass implements the operations on input data within the `forward` method.

Within `__init__` we are using PyTorch `nn.Sequential()` which is a sequential container. Modules will be added to it in the order they are passed in the constructor.

PyTorch `nn.Linear()` applies an affine linear transformation to the incoming data with the following arguments;

* `in_features`, an `int`, an integer; for example `28*28` or `512`.
* `out_features`, an `int`, an integer; for example `512` or `10`.
* `bias`, a `bool`, a boolean, true or false, defaults to `True`, we can omit (leave out) this argument in this scenario.

PyTorch `nn.ReLU()` applies the rectified linear unit function element-wise.

Let us step through this, when our `model` is run;

* The input will first be passed to `Linear(28*28, 512)`.
* The output of `Linear(28*28, 512)` will be used as the input to the first `ReLU()`.
* The output of the first `ReLU()` will become the input for `Linear(512, 512)`.
* The output of `Linear(512, 512)` will be used as the input to the second `ReLU()`.
* The output of the second `ReLU()` will become the input for `Linear(512, 10)`.

The PyTorch `to()` function within `nn.Module` will send all modules in the `nn.Sequential()` to the device (such as `"mps"` or `"cuda"`).

Finally we can `print()` our `model`.

````python
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10)
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits

model = NeuralNetwork().to(device)
print(model)
````
