### Access Valkey using Redis-rs

If you are just looking at this notebook statically in a browser or VS Code you can just read the outcome of this section.

You need to run Valkey locally if you want to run this Rust code.

To enable asynchronous clients, we have enabled the relevant feature in the `redis` Rust crate which is `tokio-comp`.

The trait `AsyncCommands` implements common redis commands over asynchronous connections. This allows you to send commands straight to a connection or client. This allows you to use nicer syntax for some common operations.

We can use this trait with `use redis::AsyncCommands;`.

All async connections are cheap to clone, and clones can be used concurrently from multiple threads.

`#[tokio::main]` marks the async function to be executed by the selected runtime.

We can define the `main()` function like so;

```rust
#[tokio::main]
async fn main() -> redis::RedisResult<()> {
    let client = redis::Client::open("redis://127.0.0.1/")?;
    let mut con = client.get_multiplexed_async_connection().await?;

    let _: () = con.set("video_narrator", "アンドリュー・ポッター | 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 Andrew Potter 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 | يوسف عبد العليم بوتر").await?;
    let val: String = con.get("video_narrator").await?;
    println!("{val}");

    Ok(())
}
```

Using `open()` within Redis-rs `Client` with the argument url; `"redis://127.0.0.1/"` does not actually open a connection to Valkey yet, but it does perform some basic checks on this url.

Using `get_multiplexed_async_connection()` returns an async connection from the client.

Note that `.await?` corresponds to the Rust keyword `await`, which suspends execution until the result of a `Future` is ready.

In this scenario `con` corresponds to a `MultiplexedConnection`, a connection object which can be cloned, allowing requests to be be sent concurrently on the same underlying connection.

Using `set()` from Redis-rs we can provide the arguments;

* `key` is `"video_narrator"`.
* `value` is `"アンドリュー・ポッター | 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 Andrew Potter 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 | يوسف عبد العليم بوتر"`.

Using `get()` from Redis-rs we can provide the argument;

* `key` is `"video_narrator"`.

The value of this key can then goes into `val`.

Finally, we can print out the `val` using `println!()`.

Within Rust, `Ok` is one of two variants of the `Result` enum. It represents success or a successful operation.

The line `main()` will call this function.

````rust
use redis::AsyncCommands;

#[tokio::main]
async fn main() -> redis::RedisResult<()> {
    let client = redis::Client::open("redis://127.0.0.1/")?;
    let mut con = client.get_multiplexed_async_connection().await?;

    let _: () = con.set("video_narrator", "アンドリュー・ポッター | 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 Andrew Potter 🏴󠁧󠁢󠁳󠁣󠁴󠁿🏴󠁧󠁢󠁥󠁮󠁧󠁿🇨🇦 | يوسف عبد العليم بوتر").await?;
    let val: String = con.get("video_narrator").await?;
    println!("{val}");

    Ok(())
}
````

````rust
main()
````

You can see that we can `set()` and `get()` the current video narrator within Valkey.
