### Import dependencies

We will import `pandas`.

```python
import pandas as pd
```

### Import csv files into a pandas `DataFrame`

We will use pandas to import two csv files. Comma-separated values files are really just text files which can store tabular data like a spreadsheet or a database.

We will use `read_csv()` within pandas to read a csv into a pandas `DataFrame`. We will call `read_csv()` with the following argument;

* `filepath_or_buffer` in this scenario, is the filename of the csv file.

A pandas `DataFrame` can be used for working with tabular data like a spreadsheet.

```python
countries_csv = "../quantumsand-project-templates-resources/countries.csv"
regions_csv = "../quantumsand-project-templates-resources/regions.csv"
```

```python
pandas_dataframe_countries = pd.read_csv(countries_csv)
pandas_dataframe_regions = pd.read_csv(regions_csv)
```

### Print the contents of the pandas DataFrames

We can print out the `pandas_dataframe_countries`.

```python
print(pandas_dataframe_countries)
```

You can see the contents of `pandas_dataframe_countries`.

We can also print out the `pandas_dataframe_regions`.

```python
print(pandas_dataframe_regions)
```

You can see the contents of `pandas_dataframe_regions`.
