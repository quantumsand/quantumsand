### Import dependencies

We will import `sklearn` and `matplotlib`.

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
```

### Load the Handwritten Digits dataset

scikit-learn has built-in datasets. We can use `load_digits()` to load the UCI ML handwritten digits dataset into `uci_ml_handwritten_digits_dataset`.

```python
uci_ml_handwritten_digits_dataset = load_digits()
```

We can learn more about this UCI dataset using Python `print()` with `uci_ml_handwritten_digits_dataset.DESCR`.

```python
print(uci_ml_handwritten_digits_dataset.DESCR)
```

You can see the dataset characteristics and references above.

We can also observe the image data using Python `print()` with `uci_ml_handwritten_digits_dataset.images`.

```python
print(uci_ml_handwritten_digits_dataset.images)
```

Finally, we can observe the `shape` of the image data by using Python `print()` with `uci_ml_handwritten_digits_dataset.images.shape`.

```python
print(uci_ml_handwritten_digits_dataset.images.shape)
```

This indicates that we have 1797 samples of handwritten digits. 

We can also see that each of these images are 8 pixels by 8 pixels. 

This means that the total number of pixels for each sample is 64 pixels.

### Render the Handwritten Digits dataset

We will use Matplotlib to render a selection of the UCI ML handwritten digits dataset.

We can render the first 256 images within the 1797 total number of images.

Using 16 by 16 `subplots()` will result in 256 total images.

We are going to display 256 of the handwritten digits dataset within the same plot. Within Matplotlib we can use `subplots()` to specify what we want. We will call `subplots()` with the following arguments;

* `nrows` is 16.
* `ncols` is 16.
* `figsize` is a keyword argument, we set to `=(15, 15)`. The (width, height) in inches.

`subplots()` returns both `fig` and `ax`, the `Figure` and the `Axes`.

* `Figure` is the top level container for all the plot elements.
* `Axes` is an object that encapsulates all of the elements of an individual subplot in a figure.

Using a Python for loop with `enumerate()` alongside `ax.flat` we can iterate using a flat 1D iterator over a 2D array of subplots.

To populate the correct subplot we can use `imshow()` within Matplotlib to display the data as an image.

The index `i` allows us to use the following syntax `uci_ml_handwritten_digits_dataset.images[i]` to obtain each image.

The index `i` also allows us to obtain the classification labels using the following syntax `uci_ml_handwritten_digits_dataset.target[i]`.

We also switch the `axis()` to `off` using `ax[i][j].axis('off')`.

Finally, we can display the classification labels using `text()` with the following arguments;

* `x` is 0.8, location x.
* `y` is 0.8, location y.
* `s` is `uci_ml_handwritten_digits_dataset.target[i]`.
* `transform` is `ax.transAxes`.
* `color` is `deepskyblue`.

The `transform` of `ax.transAxes` specifies the coordinate system being used. In this scenario we are using the axes coordinates. The positions `x` and `y` are defined relative to the axes of the subplot.

```python
fig, ax = plt.subplots(16, 16, figsize=(15, 15))

for i, ax in enumerate(ax.flat):
    ax.imshow(uci_ml_handwritten_digits_dataset.images[i], cmap='binary')
    ax.axis('off')
    ax.text(0.8, 0.8, uci_ml_handwritten_digits_dataset.target[i],
            transform=ax.transAxes, color='deepskyblue')
```

You can see 256 items from the Handwritten Digits dataset.
