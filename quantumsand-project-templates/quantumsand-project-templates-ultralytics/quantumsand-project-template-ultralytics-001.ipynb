### YOLO Object Detection

Load the pretrained model weights from the local filesystem `yolo26n.pt`. 

We will pass in the model weights to `YOLO()` and return the `model`.

We will also create an `array_of_images`, which only has one image for now.

```python
model_weights = "../quantumsand-project-templates-resources/yolo26n.pt"
model = YOLO(model_weights)
array_of_images = ["../quantumsand-project-templates-resources/andy-potter-20240905.jpg"]
```

Regarding the `array_of_images`, we can feed just one image into this array and append more images later if we wish to. You can use images on the local filesystem, or you can provide an HTTPS URL to an image from a web server.

We will pass this `array_of_images` into the `model` and return the output from the model into `results`.

We can also loop through the `results` using a for loop. The Python `enumerate()` helps us to iterate in situations where we want some kind of index and value whilst iterating over an iterable.

Using Python `print()` we can print out the values from the first iteration:
* `i` which is 0, the value of the index.
* `r` which is the `ultralytics.engine.results.Results` object, you can see the details below.

Using `save()` within `ultralytics.engine.results.Results` allows us to write a jpeg to the local filesystem.

We will pass in the argument `filename`. By using `i` within the `filename` we can generate a unique jpeg image for each step through the iteration. The first jpeg filename will be called`yolo_object_detection_results_0.jpg`.

```python
results = model(array_of_images)
```

```python
for i, r in enumerate(results):
    print(f"i: {i}")
    print(f"r: {r}")
    r.save(filename=f"../quantumsand-project-templates-resources/yolo_object_detection_results_{i}.jpg")
```

We will now display the first image from these `results`. 

In this example scenario we will load the generated `"yolo_object_detection_results_0.jpg"` image from the local filesystem into a PIL `Image`.

PIL is Pillow, a Python Imaging Library (fork). This adds image processing capabilities. In this example scenario we are just using `open()` to open and identify the given image file. We will pass in the argument;
* `fp` which is just a filename string.

We will also convert the `pil_image` into a `numpy_array`. In this scenario you do not need to create a NumPy array, but it can be useful to have the image data available as a `numpy_array` for other reasons.

Using `asarray()` within NumPy with the provided `pil_image` provides the `numpy_array`.

Finally, we can use `imshow()` within Matplotlib to render the `numpy_array` obtained from the resulting jpeg image.

```python
pil_image = Image.open("../quantumsand-project-templates-resources/yolo_object_detection_results_0.jpg", 'r')
numpy_array = np.asarray(pil_image)
imshow(numpy_array)
```

You can see the results of the YOLO Object Detection.
