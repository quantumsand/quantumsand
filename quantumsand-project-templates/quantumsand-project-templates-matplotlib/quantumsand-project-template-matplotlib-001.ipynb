## Quantum Sand Project Template Matplotlib 001

### Import dependencies

We will import `matplotlib` and `numpy`.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

### Create evenly spaced x-axis values with NumPy

We will use `arange()` within NumPy to return evenly spaced values within an interval which we can choose. We will call `arange()` with the following arguments;

* `start` is $0$.
* `stop` is $10*\pi$.
* `step` is $0.1$.

```python
x_axis_evenly_spaced_values = np.arange(0, 10*np.pi, 0.1)
```

Print out the `x_axis_evenly_spaced_values`.

```python
print(x_axis_evenly_spaced_values)
```

### Generate values for sine (sin) and cosine (cos)

```python
sine_wave_values = np.sin(x_axis_evenly_spaced_values)
```

Print out the `sine_wave_values`.

```python
print(sine_wave_values)
```

```python
cosine_wave_values = np.cos(x_axis_evenly_spaced_values)
```

Print out the `cosine_wave_values`.

```python
print(cosine_wave_values)
```

### Render the graph for sine (sin) and cosine (cos)

Create a new `figure()` with matplotlib and display the graph for sine and cosine using `plot()`.

```python
plt.figure()
```

```python
plt.plot(x_axis_evenly_spaced_values, sine_wave_values, color='pink')
plt.plot(x_axis_evenly_spaced_values, cosine_wave_values, color='deepskyblue')
```

You can see the pink sine wave and the blue cosine wave.
