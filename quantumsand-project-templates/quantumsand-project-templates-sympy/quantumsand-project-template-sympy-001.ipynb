### Solve a system of linear equations for two vehicles using SymPy

A Volkswagen ID. Buzz departs from Madinah to Makkah.

A Porsche Taycan, 30 minutes later, also departs from Madinah to Makkah.

The distance between Makkah and Madinah is 435km.

The Volkswagen ID. Buzz is travelling at 72km/hour (45mph) or 1.2 kilometres per minute; for a total journey time of 6 hours.

The Porsche Taycan is travelling at 96km/hour (60mph) or 1.6 kilometres per minute; for a total journey time of 4.5 hours. 

We can express the distance travelled by the ID. Buzz in km as; $$ d_b = 1.2t $$

We can express the distance travelled by the Porsche Taycan in km as; $$ d_p = 1.6 (t - 30) $$

Solve the distance and time, for the Porsche Taycan catching up with the ID. Buzz.

SymPy reuses Python syntax, so we have to define the symbols. Using `symbols()` from SymPy;

* `d` is the distance.
* `t` is the time.

Using `Eq()` from SymPy we can express the distance travelled by both vehicles;

* the distance for the ID. Buzz in km $ d_b = 1.2t $ can be expressed as `Eq(d, 1.2*t)` as `eq1`.
* the distance for the Porsche Taycan in km $ d_p = 1.6 (t - 30) $ can be expressed as `Eq(d, 1.6*(t - 30))` as `eq2`.

We can find the solution using `solve()` from SymPy with the following arguments;

* `f` the expressions for both `eq1` and `eq2` which we can pass in as `[eq1,eq2]`.
* `symbols` the symbols we defined earlier, `d` and `t`, which we can pass in as `(d, t)`.

This algebraically solves equations and systems of equations into the `solution`.

This `solution` is: `{d: 144.000000000000, t: 120.000000000000}`.

The solution for the Porsche Taycan catching up with the ID. Buzz is;

* distance `d` is 144 km.
* time `t` is 120 minutes.

We can access the values from `solution` using the following syntax;

* distance is `solution[d]`.
* time is `solution[t]`.

```python
d, t = sp.symbols('d t')

eq1 = sp.Eq(d, 1.2*t)
eq2 = sp.Eq(d, 1.6*(t - 30))

solution = sp.solve([eq1,eq2], (d, t))
print("print() solve([eq1,eq2], (d, t))")
print(solution)
print()

print("print() solution[d]")
print(solution[d])
print()

print("print() solution[t]")
print(solution[t])
```

We can also generate a plot with Matplotlib to demonstrate that these values are correct.

### Generate one thousand evenly spaced time values using NumPy

We are going to plot the time and distance for both the ID. Buzz and the Porsche Taycan.

We know that the solution time is 120 minutes and the solution distance is 144 km.

We can therefore specify the maximum values for time and distance to plot;

* `max_time_in_minutes` is `140`.
* `max_distance_in_km` is `160`.

Finally, we can generate one thousand evenly spaced time values using `linspace()` from NumPy.

Using these arguments;

* `start` is $0$.
* `stop` is `max_time_in_minutes` or $140$.
* `num` is $1000$.

We can put these values into `evenly_spaced_time_values`.

```python
max_time_in_minutes = 140
max_distance_in_km = 160

evenly_spaced_time_values = np.linspace(0, max_time_in_minutes, 1000)
```

Print out the `evenly_spaced_time_values`.

```python
print(evenly_spaced_time_values)
```

### Render the distance-time graph for the vehicles

Generate values for the distance travelled by the ID. Buzz `d_buzz` and the Porsche Taycan `d_porsche`.

We can now generate the values we need for plotting the distance of both vehicles.

```python
d_buzz = 1.2 * evenly_spaced_time_values
d_porsche = 1.6 * (evenly_spaced_time_values - 30)
```

Create a new `figure()` with Matplotlib.

We can create a figure with a single subplot; one row and one column using; `fig, ax = plt.subplots()`.

Use `title()` to define the plot title `'The Porsche Taycan catches up with the ID. Buzz'`.

Use `xlabel()` to define the x-axis label `'time in minutes'`.

Use `ylabel()` to define the y-axis label `'distance in km'`.

Use `set_xlim()` to set the x-axis view limits, `[0, max_time_in_minutes]`.

Use `set_ylim()` to set the y-axis view limits, `[0, max_distance_in_km]`.

Use `plot()` to plot the ID. Buzz distance.
`x` is `evenly_spaced_time_values`, `y` is `d_buzz` and colour `c` is `'pink'`.

Use `plot()` to plot the Porsche Taycan distance.
`x` is `evenly_spaced_time_values`, `y` is `d_porsche` and colour `c` is `'deepskyblue'`.

Finally, we can plot the solutions from SymPy as lines on the plot.

Use `axvline()` to draw a vertical line across an entire plot.
`x` is `solution[t]`, `color` is `'lightseagreen'` and `linestyle` is `'--'`.

Use `axhline` to draw a horizontal line across an entire plot.
`y` is `solution[d]`, `color` is `'lightseagreen'` and `linestyle` is `'--'`.

```python
plt.figure()
```

```python
fig, ax = plt.subplots()

plt.title('The Porsche Taycan catches up with the ID. Buzz')
plt.xlabel('time in minutes')
plt.ylabel('distance in km')

ax.set_xlim([0, max_time_in_minutes])
ax.set_ylim([0, max_distance_in_km])

plt.plot(evenly_spaced_time_values, d_buzz, c='pink')
plt.plot(evenly_spaced_time_values, d_porsche, c='deepskyblue')

plt.axvline(x=solution[t], color='lightseagreen', linestyle='--')
plt.axhline(y=solution[d], color='lightseagreen', linestyle='--')
```

You can see the pink line for the ID. Buzz and the blue line for the Porsche Taycan.

You can also see that the solutions for the time and distance from SymPy `solve()` are correct.
