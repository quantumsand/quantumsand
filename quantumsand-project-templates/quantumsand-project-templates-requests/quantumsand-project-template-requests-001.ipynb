### Running Quantum Sand Rails locally

If you are just looking at this notebook statically in a browser or VS Code you can just read this notebook without running the Python code.

Having the Quantum Sand Rails; Ruby-on-Rails PostGIS API; running locally is necessary for running the Python code later in this notebook.

The `seeds.rb` within Quantum Sand Rails has some test data which we can use for this example;

This block of Ruby code contains some seed data for the `geospatial_traces` table;

```ruby
geospatial_traces = [
  {
    name: "Harbour seals in Christchurch Bay",
    geospatial: "POINT(50.713294760937224 -1.6601009519012668)",
    data: {
      name: "Harbour seal",
      scientific_name: "Phoca vitulina"
    }
  }
]

geospatial_traces.each do |geospatial_trace|
  GeospatialTrace.where(name: geospatial_trace[:name]).first_or_create do |trace|
    trace.name = geospatial_trace[:name]
    trace.geospatial = geospatial_trace[:geospatial]
    trace.data = geospatial_trace[:data]

    puts "Created geospatial_trace with name: #{geospatial_trace[:name]}"
  end
end
```

The `geospatial_traces` have the following attributes; the columns in the database table;

* `name` is a string, `"Harbour seals in Christchurch Bay"`.
* `geospatial` is PostGIS geometry, `"POINT(50.713294760937224 -1.6601009519012668)"`.
* `data` is JSON, `{name: "Harbour seal", scientific_name: "Phoca vitulina"}.`

We will use Python with Requests to `get()` this data from the `geospatial_traces` database table within the Rails app.

### Access the Rails PostGIS API over HTTP using Requests

If you are just looking at this notebook statically in a browser or VS Code you can just read the outcome of this section.

You need to run Quantum Sand Rails locally if you want to run this Python code.

We can define `rails_postgis_api_url` as `"http://127.0.0.1:3000/geospatial_traces"`.

This is the url within Quantum Sand Rails corresponding to the `geospatial_traces`.

Using `get()` within Requests sends a GET request and returns a `Response` object. 

We can provide `get()` with the following argument;

* `url` is `rails_postgis_api_url`.

We can store the response object in `rails_postgis_api_response`.

We can decode the JSON response body as a Python object using `json()` from Requests into `rails_postgis_api_response_json`.

We can also use `dumps()` from Python json to change the JSON response body Python object into a string `json_string`.

Finally, `rich.console.Console()` provides a high-level console interface with finer control and advanced terminal rendering including colour systems.

We can now use `Console()` and `print_json()` with the argument `json_string`. 

This provides a “pretty-print” JSON output, formatting the JSON for improved readability with colourisation.

````python
rails_postgis_api_url = "http://127.0.0.1:3000/geospatial_traces"

rails_postgis_api_response = requests.get(rails_postgis_api_url)

print("print() rails_postgis_api_response")
print(rails_postgis_api_response)
print()

print("print() rails_postgis_api_response.json()")
rails_postgis_api_response_json = rails_postgis_api_response.json()
print(rails_postgis_api_response_json)
print()

print("Console() print_json() json_string")
json_string = json.dumps(rails_postgis_api_response_json)

console = Console()
console.print_json(json_string)
````

You can see the same data that was used to seed the `geospatial_traces` table.
