### Import dependencies

We will import `pydicom` and `matplotlib`

```python
import pydicom
```

```python
import matplotlib.pyplot as plt
```

### Read the cancer imaging DICOM file



```python
cancer_imaging_dicom = pydicom.dcmread("../quantumsand-project-templates-resources/ID_0000_AGE_0060_CONTRAST_1_CT.dcm")
```

Print out the `cancer_imaging_dicom` file.

```python
print(cancer_imaging_dicom)
```

### Extract and render the cancer imaging DICOM image

```python
cancer_imaging_dicom_pixel_array = cancer_imaging_dicom.pixel_array
```

Create a new `figure()` with matplotlib and display the pixel array using `imshow()`.

```python
plt.figure()
```

```python
plt.imshow(cancer_imaging_dicom_pixel_array, cmap="bone")
```

You can see the cancer imaging scan.
